# Which Algeria player had the highest overall rating?

```python
algeria_players = df[df['Nationality'] == 'Algeria']
top_10_algeria_players = algeria_players[['Name', 'Overall']].sort_values(by='Overall', ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x='Overall', y='Name', data=top_10_algeria_players,color='red')
plt.title('Top 10 Algerian Players by Overall Rating')
plt.xlabel('Overall Rating')
plt.ylabel('Player Name')
plt.show()
```

# Which back had the highest rating for ‚ÄòSliding tackle‚Äô?



```python
highest_sliding_tackle = df[['Name', 'Sliding tackle']].sort_values(by='Sliding tackle', ascending=False).head(10)
plt.figure(figsize=(8, 4))
sns.barplot(x='Sliding tackle', y='Name', data=highest_sliding_tackle, color='blue')
plt.title('Player with Highest Sliding Tackle Rating')
plt.xlabel('Sliding Tackle Rating')
plt.ylabel('Player Name')
plt.show()
```

# Which preferred position type of England has on average the highest overall rating?

```python
england_players = df[df['Nationality'] == 'England']
average_overall_by_position = england_players.groupby('Preferred Positions Type')['Overall'].mean().reset_index()
highest_avg_position = average_overall_by_position.sort_values(by='Overall', ascending=False).head(1)
plt.figure(figsize=(10, 6))
sns.barplot(x='Overall', y='Preferred Positions Type', data=average_overall_by_position, color='green')
plt.title('Average Overall Rating by Preferred Position Type (England Players)')
plt.xlabel('Average Overall Rating')
plt.ylabel('Preferred Position Type')
plt.axvline(x=highest_avg_position['Overall'].values[0], color='red', linestyle='--', label='Highest Avg Position')
plt.legend()
plt.show()
```

# Brazil‚Äôs forwards have a higher average overall rating than the backs


```python
brazil_players = df[df['Nationality'] == 'Brazil']
forwards_brazil = brazil_players[brazil_players['Preferred Positions Type'].str.contains('Forward', na=False)]
defenders_brazil = brazil_players[brazil_players['Preferred Positions Type'].str.contains('Back', na=False)]
average_forward_rating = forwards_brazil['Overall'].mean()
average_defender_rating = defenders_brazil['Overall'].mean()
comparison_data = pd.DataFrame({
    'Position': ['Forwards', 'Defenders'],
    'Average Rating': [average_forward_rating, average_defender_rating]
})


plt.figure(figsize=(8, 8))
plt.pie(comparison_data['Average Rating'], labels=comparison_data['Position'], autopct='%1.1f%%', startangle=90, colors=['#1f77b4', '#ff7f0e'])
plt.title('Average Rating Comparison: Forwards vs. Defenders (Brazil Players)')
plt.axis('equal')
plt.show()
```

# Which country has the oldest player?



```python
oldest_player = df.loc[df['Age'].idxmax()]
oldest_player_country = oldest_player['Nationality']
oldest_player_age = oldest_player['Age']
data = {
    'Country': [oldest_player_country],
    'Age': [oldest_player_age]
}
visualization_df = pd.DataFrame(data)


plt.figure(figsize=(6, 4))
plt.bar(visualization_df['Country'], visualization_df['Age'], color='orange')
plt.title('Oldest Player by Country')
plt.xlabel('Country')
plt.ylabel('Age')
plt.ylim(0, 50)
plt.xticks(rotation=45)
plt.show()
```

# Which player from Portugal, that is younger than 25, has the highest overall rating?


```python
portugal_young_players = df[(df['Nationality'] == 'Portugal') & (df['Age'] < 25)]
highest_rated_young_player = portugal_young_players.loc[portugal_young_players['Overall'].idxmax()]
player_name = highest_rated_young_player['Name']
overall_rating = highest_rated_young_player['Overall']
data = {
    'Player': [player_name],
    'Overall Rating': [overall_rating]
}

visualization_df = pd.DataFrame(data)
plt.figure(figsize=(6, 4))
plt.bar(visualization_df['Player'], visualization_df['Overall Rating'], color='blue')
plt.title('Highest Rated Young Player from Portugal (< 25 years old)')
plt.xlabel('Player')
plt.ylabel('Overall Rating')
plt.ylim(0, 100)
plt.xticks(rotation=45)
plt.show()
```
